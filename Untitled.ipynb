```python
def catconvert(df1):
    df_Sex = pd.get_dummies(df1['Sex'],prefix="Sex")
    df_Emb = pd.get_dummies(df1['Embarked'],prefix="Emb")
    df1 = pd.concat([df1, df_Sex, df_Emb], axis=1)
    del df1["Sex"]
    del df1["Embarked"]

    return df1
df_train=catconvert(df_train)
```
